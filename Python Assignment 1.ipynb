### 2. Create the below pattern using nested for loop in Python. 
 
### *  
### * *  
### * * *  
### * * * *  
### * * * * *  
### * * * *  
### * * *  
### * *  
### * 

```python
n=5;
for i in range(n):
    for j in range(i):
        print ('* ', end = "")
    print('')
for i in range(n,0,-1):
    for j in range(i):
        print('* ', end="")
    print('')
```
